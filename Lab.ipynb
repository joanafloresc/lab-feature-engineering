Identify columns that over 85% missing values\

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
perc_nulls = data.isna().sum()/len(data)
drop_list = list(perc_nulls[perc_nulls > 0.85].index)
drop_list.append('OSOURCE')
drop_list.append('ZIP')
drop_list
```

```python
data_new = data.drop(drop_list, axis=1)
data_new.head()
```

```python
cat = data_new.select_dtypes('object')
cat.head()
```

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
cat['GENDER'].value_counts().index
```

```python
cat['GENDER'] = cat['GENDER'].apply(lambda x: 'F' if x=='F' else ('M' if x=='M' else 'other'))
```

### WEALTH1

```python
data_new['WEALTH1'].value_counts(dropna=False)/len(num)
```

```python
list(data_new.columns)
```

```python
list((num.isna().sum()/len(num) > 0).index).remove('WEALTH1')
list((num.isna().sum()/len(num) == 0).index)
```

```python
to_remove = list((num.isna().sum()/len(num) > 0).index)
to_remove.remove('WEALTH1')
to_remove
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer() #default = 5 neighbours

x_wealth = num.drop(to_remove, axis=1)
x_wealth_np = imputer.fit_transform(x_wealth)
x_wealth_df = pd.DataFrame(x_wealth_np, columns=x_wealth.columns, index=x_wealth.index)
x_wealth_df.dtypes
```

```python
data_new['WEALTH1'] = x_wealth_df['WEALTH1'].apply(lambda x: int(round(x)))
data_new['WEALTH1'].value_counts()
```
